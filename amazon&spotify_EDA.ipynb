#Q-3)  What is the distribution of discounted prices vs. actual prices?

Ans)

```python
df["discounted_price"] = pd.to_numeric(df["discounted_price"].str.replace(',', '').str.replace('₹', ''), errors="coerce")
df["actual_price"] = pd.to_numeric(df["actual_price"].str.replace(',', '').str.replace('₹', ''), errors="coerce")

plt.figure(figsize=(10, 6))
plt.scatter(df["actual_price"], df["discounted_price"], alpha=0.5)
plt.xlabel("Actual Price (₹)")
plt.ylabel("Discounted Price (₹)")
plt.title("Distribution of Discounted Prices vs. Actual Prices")
plt.grid()
plt.show()
```

# Q-4) How does the average discount percentage vary across categories?

Ans)

```python

df["discount_percentage"] = pd.to_numeric(df["discount_percentage"].str.replace('%', ''), errors="coerce")

# Group by category and calculate the average discount percentage
avg_discount_by_category = df.groupby("category")["discount_percentage"].mean().reset_index()

# Sort categories by average discount percentage
avg_discount_by_category = avg_discount_by_category.sort_values(by="discount_percentage", ascending=False)

# Plot the average discount percentage by category using a different plot type
plt.figure(figsize=(12, 6))
plt.plot(avg_discount_by_category["category"], avg_discount_by_category["discount_percentage"], marker='o', linestyle='-', color='b')
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Average Discount Percentage")
plt.title("Average Discount Percentage by Category")
plt.grid()
plt.show()
```

#Q-5) What are the most popular product names?

Ans)

```python
most_popular_products = df.groupby("product_name")["rating_count"].sum().reset_index()
most_popular_products = most_popular_products.sort_values(by="rating_count", ascending=False)
print(most_popular_products.head(10))
```

# Q-6) What are the most popular product keywords?

Ans)

#Q-7) What are the most popular product reviews?

# Q-8) What is the correlation between discounted_price and rating?

Ans)

```python
df["discounted_price"] = df["discounted_price"].astype(str).str.replace(',', '').str.replace('₹', '')

df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors="coerce")
correlation = df["discounted_price"].corr(df["rating"])

print(f"Correlation between Discounted Price and Rating: {correlation}")
```

# Q-9) What are the Top 5 categories based on the highest ratings?

Ans)

```python
top_categories = df.groupby("category")["rating"].mean().reset_index()

top_categories = top_categories.sort_values(by="rating", ascending=False).head(5)

print("Top 5 Categories Based on Highest Ratings:")
print(top_categories)
```

# Q-10)  Identify any potential areas for improvement or optimization based on the data analysis?

Ans)

```python
if "discount_percentage" in df.columns:
    df["discount_percentage"] = pd.to_numeric(df["discount_percentage"].astype(str).str.replace('%', ''), errors="coerce")

# 1️ Identify Low-Rated Categories
low_rated_categories = df.groupby("category")["rating"].mean().sort_values().head(5)
print("\n🔹 Categories with the Lowest Ratings:")
print(low_rated_categories)

# 2️ Find Products with High Discounts but Low Ratings
low_rating_high_discount = df[(df["rating"] < 3) & (df.get("discount_percentage", 0) > 30)]
print("\n🔹 Products with High Discounts but Low Ratings:")
print(low_rating_high_discount[["product_name", "rating", "discount_percentage"]].head(10))

# 3️ Identify Popular Products with Few Reviews
popular_but_few_reviews = df[(df["rating"] > 4) & (df["rating_count"] < 50)]
print("\n🔹 Popular Products with Few Reviews:")
print(popular_but_few_reviews[["product_name", "rating", "rating_count"]].head(10))

# 4️ Identify Potentially Overpriced Products
high_price_products = df[df["discounted_price"] > df["actual_price"] * 0.9]  # Less than 10% discount
print("\n🔹 Potentially Overpriced Products:")
print(high_price_products[["product_name", "actual_price", "discounted_price"]].head(10))

# 5️ Identify High-Rated Products with Low Sales
high_rating_low_sales = df[(df["rating"] > 4) & (df["rating_count"] < 100)]
print("\n🔹 High-Rated Products with Low Sales:")
print(high_rating_low_sales[["product_name", "rating", "rating_count"]].head(10))
```

```python
df2["Duration (s)"] = df2["Duration (ms)"] / 1000
df2.drop(columns=["Duration (ms)"], inplace=True)
```

# Q-1)  Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful?

Ans)

```python
if df2.isnull().sum() > 0:
    df2.dropna(inplace=True)
    print("Null values found and removed.")
else:
    print("No null values found.")

if df2.duplicated().sum() > 0:
    df2.drop_duplicates(inplace=True)
    print("Duplicate rows found and removed.")
else:
    print("No duplicate rows found.")

print(df2.info())
print(df2.head())
```

# Q-2)  What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram?

Ans)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df2["Popularity"], bins=20, kde=True, color="blue")

plt.xlabel("Popularity Score")
plt.ylabel("Count of Tracks")
plt.title("Distribution of Popularity Among Tracks")
plt.show()
```

# Q-3)  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?

Ans)

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df2["Duration (s)"], y=df2["Popularity"], alpha=0.7, color="blue")

plt.xlabel("Duration (Seconds)")
plt.ylabel("Popularity Score")
plt.title("Relationship Between Track Duration and Popularity")
plt.show()
```

# Q-4) Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot?

Ans)

```python
artist_counts = df2["Artist"].value_counts()
top_artists = artist_counts.head(10)

plt.style.use("ggplot")

plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.index, y=top_artists.values, palette="viridis")

plt.xlabel("Artist Name")
plt.ylabel("Number of Tracks")
plt.title("Top 10 Artists with the Most Tracks in the Dataset")
plt.xticks(rotation=45)
plt.show()
```

# Q-5) What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each?

Ans)

```python
least_popular_tracks = df2.nsmallest(5, "Popularity")[["Artist", "Track Name", "Popularity"]]

print("Top 5 Least Popular Tracks:")
print(least_popular_tracks)
```

# Q-6) Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist?

Ans)

```python
artist_avg_popularity = df2.groupby("Artist")["Popularity"].mean().sort_values(ascending=False)

top_5_popular_artists = artist_avg_popularity.head(5)

print("Top 5 Most Popular Artists Based on Average Popularity:")
print(top_5_popular_artists)
```

# Q-7)  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?

Ans)

```python
artist_avg_popularity = df2.groupby("Artist")["Popularity"].mean().sort_values(ascending=False)

top_5_artists = artist_avg_popularity.head(5).index

most_popular_tracks = df2.loc[df2.groupby("Artist")["Popularity"].idxmax(), ["Artist", "Track Name", "Popularity"]]
most_popular_tracks = most_popular_tracks[most_popular_tracks["Artist"].isin(top_5_artists)]

print("Most Popular Track for Each of the Top 5 Most Popular Artists:")
print(most_popular_tracks)
```

# Q-8) Visualize relationships between multiple numerical variables simultaneously using a pair plot?

Ans)

```python
numerical_cols = ["Popularity", "Duration (s)"]

sns.pairplot(df2[numerical_cols], diag_kind="kde", markers="o", corner=True)
plt.show()
```

# Q-9) Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot?


Ans)

```python
top_10_artists = df2["Artist"].value_counts().head(10).index
df_top_artists = df2[df2["Artist"].isin(top_10_artists)]

plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)

sns.boxplot(x=df_top_artists["Artist"], y=df_top_artists["Duration (s)"], palette="Set2")
plt.xlabel("Artist")
plt.ylabel("Track Duration (Seconds)")
plt.title("Distribution of Track Durations Across Top 10 Artists")
plt.show()
```

# Q-10) How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot?

Ans)

```python
top_10_artists = df2["Artist"].value_counts().head(10).index
df_top_artists = df2[df2["Artist"].isin(top_10_artists)]

plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)

sns.violinplot(x=df_top_artists["Artist"], y=df_top_artists["Popularity"], palette="Set2", inner="quartile")
plt.xlabel("Artist")
plt.ylabel("Popularity Score")
plt.title("Distribution of Track Popularity Across Top 10 Artists")
plt.show()
```
