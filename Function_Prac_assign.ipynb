# Q-1)  Write a Python function that takes a list of numbers as input  and returns the sum of all even numbers in the list.

Ans)

```python
def sum_of_even_numbers(num):
    lis = []
    total_sum = 0
    for i in num:
        if i % 2 == 0:
            total_sum += i
            lis.append(i)

    return f" 'Sum of Even number`s {lis} are' => {total_sum}"

n = input("Enter number`s seperated by a comma : ")
num = list(map(int,n.split(",")))

sum_of_even_numbers(num)
```

# Q-3)  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

Ans)

```python
def square_of_number(num):
    lis = []
    for i in num:
        lis.append(i**2)

    return f"'Square of a given list is' => {lis}"

n = input("Enter number`s seperated by a comma : ")
num = list(map(int,n.split(",")))

square_of_number(num)
```

#Q-4)  Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans)

```python
def is_prime(num):

    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):  # Check divisors up to âˆšnum
        if num % i == 0:
            return False
    return True

def primes_in_range(start, end):

    primes = [n for n in range(start, end + 1) if is_prime(n)]
    return primes


prime_numbers = primes_in_range(1, 200)
print("Prime numbers from 1 to 200:", prime_numbers)
```


#Q-5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

Ans)

```python
def fibonacci(n):

    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


n = int(input("Enter the number of terms  you want to see in the Fibonacci sequence: "))
fib_generator = fibonacci(n)

for number in fib_generator:
    print(number, end=" ")
```


# Q-6) Write a generator function in Python that yields the powers of 2 up to a given exponent.

Ans)

```python
def powers_of_two(n):

    for i in range(n + 1):
        yield 2 ** i


n = int(input("Enter the exponent for the power of 2: "))
powers_generator = powers_of_two(n)


powers_list = list(powers_generator)
print("Powers of 2 up to 2^", n, ":", powers_list)
```

# Q-7)  Implement a generator function that reads a file line by line and yields each line as a string.

Ans)

```python
def read_str_line(input_str):

    for line in input_str.splitlines():
        yield line  # Yield each line one by one

# Example usage
input_str = """Line 1: Hello, World!
Line 2: Welcome to Python Generators.
Line 3: This is a multiline string."""

for line in read_str_line(input_str):
    print(line)
```

# Q-8)  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

Ans)

```python

data = [(1, 3), (4, 1), (6, 9), (3, 2)]

# Sorting based on the second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

print("Sorted list:", sorted_data)
```

# Q-9) Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Ans)

```python

celsius_temps = [0, 20, 30, 40, 100]

fahrenheit_temps = list(map(lambda c: c * 9 / 5 + 32, celsius_temps))


print("Temperatures in Celsius:", celsius_temps)
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

# Q-10) Create a Python program that uses `filter()` to remove all the vowels from a given string.

Ans)

```python
input_string = "I`m pursuing Data science course in PW"

vowels = {'a', 'e', 'i', 'o', 'u'}

# Use filter() with a lambda function to remove vowels
result = ''.join(filter(lambda char: char.lower() not in vowels, input_string))

print("Original string:", input_string)
print("String without vowels:", result)
```
