# Q-1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

Ans)

```python
import numpy as np
from scipy.stats import norm

sample_data = [100, 102, 98, 101, 99, 97, 105, 103, 106, 100]

# Known population mean
mu = 100
sigma = 10

n = len(sample_data)

sample_mean = np.mean(sample_data)
std_error = sigma / np.sqrt(n)

z_score = (sample_mean - mu) / std_error

alpha = 0.05

critical_z = norm.ppf(1 - alpha / 2)

print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"Critical Z-Value: {critical_z}")

if abs(z_score) > critical_z:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

# Q-4)  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

Ans)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_data = [98, 102, 100, 105, 97, 104, 103, 101, 99, 100]

mu = 100
sigma = 10

n = len(sample_data)
sample_mean = np.mean(sample_data)

std_error = sigma / np.sqrt(n)

z_score = (sample_mean - mu) / std_error

alpha = 0.05

critical_z = norm.ppf(1 - alpha / 2)

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma / np.sqrt(n))

plt.plot(x, y, label="Normal Distribution", color="blue")
plt.fill_between(x, y, where=(x < mu - critical_z * std_error) | (x > mu + critical_z * std_error),
                 color="red", alpha=0.5, label="Rejection Region")
plt.axvline(mu, color='black', linestyle='--', label="Population Mean (mu)")
plt.axvline(sample_mean, color='green', linestyle='--', label="Sample Mean")
plt.axvline(mu - critical_z * std_error, color='orange', linestyle='--', label="Critical Z-Value")
plt.axvline(mu + critical_z * std_error, color='orange', linestyle='--')

plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"Critical Z-Value: {critical_z}")

if abs(z_score) > critical_z:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

# Q-5)  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

Ans)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_errors(mu, sigma, n, alpha, alternative_mu):
    null_data = np.random.normal(mu, sigma, n)
    null_mean = np.mean(null_data)

    # Simulate data for the alternative hypothesis (H1)
    alt_data = np.random.normal(alternative_mu, sigma, n)
    alt_mean = np.mean(alt_data)

    std_error = sigma / np.sqrt(n)
    critical_z = norm.ppf(1 - alpha / 2)

    z_null = (null_mean - mu) / std_error
    z_alt = (alt_mean - mu) / std_error

    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, mu, sigma / np.sqrt(n))

    plt.plot(x, y, label="Null Hypothesis Distribution", color="blue")
    plt.fill_between(x, y, where=(x < mu - critical_z * std_error) | (x > mu + critical_z * std_error),
                     color="red", alpha=0.5, label="Rejection Region")

    plt.axvline(mu, color='black', linestyle='--', label="Population Mean")
    plt.axvline(null_mean, color='green', linestyle='--', label="Sample Mean (H0)")
    plt.axvline(alt_mean, color='orange', linestyle='--', label="Sample Mean (H1)")

    plt.axvline(mu - critical_z * std_error, color='purple', linestyle='--', label="Critical Z-Value")
    plt.axvline(mu + critical_z * std_error, color='purple', linestyle='--')

    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Z-Score (Null Hypothesis, H0): {z_null}")
    print(f"Z-Score (Alternative Hypothesis, H1): {z_alt}")

    if abs(z_null) > critical_z:
        print("Reject the null hypothesis (No Type 1 Error).")
    else:
        print("Fail to reject the null hypothesis (Type 1 Error).")

    if abs(z_alt) > critical_z:
        print("Reject the null hypothesis (No Type 2 Error).")
    else:
        print("Fail to reject the null hypothesis (Type 2 Error).")

# Call the function with example values
plot_errors(mu=100, sigma=10, n=100, alpha=0.05, alternative_mu=105)

```

# Q-8) Simulate data and perform both Z-test and T-test, then compare the results using Python.

Ans)

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample_data)
    std_error = population_std / np.sqrt(sample_size)
    z_stat = (sample_mean - population_mean) / std_error
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    return z_stat, p_value_z

def t_test(sample_data, population_mean, sample_size):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    std_error = sample_std / np.sqrt(sample_size)
    t_stat, p_value_t = stats.ttest_1samp(sample_data, population_mean)
    return t_stat, p_value_t

def simulate_and_compare_tests():
    np.random.seed(42)
    sample_data = np.random.normal(loc=100, scale=15, size=30)

    population_mean = 100
    population_std = 15
    sample_size = len(sample_data)

    z_stat, p_value_z = z_test(sample_data, population_mean, population_std, sample_size)
    t_stat, p_value_t = t_test(sample_data, population_mean, sample_size)

    print(f"Z-Test Results:")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"Z-test P-value: {p_value_z:.4f}")

    print(f"\nT-Test Results:")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"T-test P-value: {p_value_t:.4f}")

    plt.figure(figsize=(10, 6))

    plt.hist(sample_data, bins=10, alpha=0.7, color='blue', label='Sample Data')

    plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')

    plt.title("Histogram of Sample Data with Population Mean")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

simulate_and_compare_tests()
```

# Q-9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

Ans)

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    std_error = sample_std / np.sqrt(sample_size)

    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    margin_of_error = z_value * std_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Standard Deviation: {sample_std}")
    print(f"Standard Error: {std_error}")
    print(f"Z-value (for {confidence_level*100}% CI): {z_value}")
    print(f"Confidence Interval: ({lower_bound}, {upper_bound})")

    return lower_bound, upper_bound

sample_data = [85, 88, 92, 94, 90, 91, 93, 89, 87, 93]
confidence_interval(sample_data, confidence_level=0.95)
```

# Q-10) Write a Python program to calculate the margin of error for a given confidence level using sample data.

Ans)

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence_level):
    n = len(sample)
    mean = np.mean(sample)
    std_dev = np.std(sample, ddof=1)

    alpha = 1 - confidence_level
    z_score = stats.norm.ppf(1 - alpha / 2)

    me = z_score * (std_dev / np.sqrt(n))
    return me

sample_data = [10, 12, 14, 15, 13, 12, 14, 16, 15, 13]
confidence_level = 0.95

me = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error: {me:.4f}")
```

# Q-11)  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

Ans)

```python
def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    prior_not_A = 1 - prior_A
    marginal_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)

    posterior_A_given_B = (likelihood_B_given_A * prior_A) / marginal_B
    return posterior_A_given_B

prior_disease = 0.01
likelihood_positive_given_disease = 0.95
likelihood_positive_given_no_disease = 0.05

posterior_probability = bayes_theorem(prior_disease, likelihood_positive_given_disease, likelihood_positive_given_no_disease)

print(f"Probability of having the disease given a positive test: {posterior_probability:.4f}")
```

# Q-14)  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

Ans)

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value


observed_data = [50, 30, 20]

total = sum(observed_data)
expected_data = [total / len(observed_data)] * len(observed_data)

chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is different from expected.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
```

# Q-16)  Implement an F-test using Python to compare the variances of two random samples.

Ans)

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        F_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        F_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * (1 - stats.f.cdf(F_stat, dfn, dfd))  # Two-tailed test

    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

# Perform the F-test
f_test(sample1, sample2)
```

# Q-18) Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def perform_anova_and_plot(*groups):

    F_stat, p_value = stats.f_oneway(*groups)

    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: At least one group mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in group means.")

    data = []
    for i, group in enumerate(groups):
        for value in group:
            data.append({"Group": f"Group {i+1}", "Value": value})

    df = pd.DataFrame(data)

    plt.figure(figsize=(8, 5))
    sns.boxplot(x="Group", y="Value", data=df, palette="Set2")
    plt.title("Comparison of Group Means (One-Way ANOVA)")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

perform_anova_and_plot(group1, group2, group3)
```

# Q-19) Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):

    print("### Checking ANOVA Assumptions ###\n")

    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: Shapiro-Wilk Test p-value = {p_value:.4f}")
        if p_value < 0.05:
            print("  -> Data is NOT normally distributed.")
        else:
            print("  -> Data appears normally distributed.")

    print("\n### Homogeneity of Variance ###")
    stat, p_value = stats.levene(*groups)
    print(f"Levenes Test p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  -> Variances are NOT equal across groups.")
    else:
        print("  -> Variances are equal across groups.")

    plt.figure(figsize=(10, 5))

    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=f'Group {i+1}', alpha=0.6, bins=10)

    plt.title("Histogram of Data (Checking Normality)")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(8, 5))
    sns.boxplot(data=groups)
    plt.title("Boxplot of Groups (Checking Homogeneity of Variance)")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

check_anova_assumptions(group1, group2, group3)
```

# Q-21)  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_f_distribution(df1, df2, x_max=5):

    x = np.linspace(0, x_max, 500)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"F({df1}, {df2})", color='blue')
    plt.fill_between(x, y, alpha=0.2, color='blue')

    alpha = 0.05
    f_critical = stats.f.ppf(1 - alpha, df1, df2)
    plt.axvline(f_critical, color='red', linestyle='dashed', label=f"Critical Value (=0.05): {f_critical:.2f}")

    plt.title("F-Distribution")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Critical Value at =0.05: {f_critical:.2f}")

plot_f_distribution(df1=5, df2=10)
```

# Q-24) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

population_variance = 100
sample_size = 30

data = np.random.normal(0, np.sqrt(population_variance), sample_size)
sample_variance = np.var(data, ddof=1)

chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

df = sample_size - 1

p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

print(f"Chi-square statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

x = np.linspace(0, 30, 100)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, label=f"Chi-square Distribution (df={df})")
plt.axvline(x=chi_square_statistic, color='r', linestyle='--', label=f"Chi-square Statistic = {chi_square_statistic:.2f}")
plt.title('Chi-Square Distribution and Test Statistic')
plt.legend()
plt.show()
```

# Q-25) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

Ans)

```python
import numpy as np
import scipy.stats as stats

successes_group1 = 80
n1 = 200
successes_group2 = 60
n2 = 150

p1 = successes_group1 / n1
p2 = successes_group2 / n2

p = (successes_group1 + successes_group2) / (n1 + n2)

z_statistic = (p1 - p2) / np.sqrt(p * (1 - p) * (1 / n1 + 1 / n2))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

# Q-26)  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


np.random.seed(0)
group1 = np.random.normal(loc=0, scale=5, size=100)
group2 = np.random.normal(loc=0, scale=10, size=100)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

F_statistic = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(group1) - 1
df2 = len(group2) - 1

p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")


plt.figure(figsize=(12, 6))
plt.hist(group1, bins=20, alpha=0.5, label="Group 1 (Var = {:.2f})".format(var1), color='blue')
plt.hist(group2, bins=20, alpha=0.5, label="Group 2 (Var = {:.2f})".format(var2), color='green')
plt.legend()
plt.title("Histogram of Group 1 and Group 2")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot([group1, group2], labels=["Group 1", "Group 2"])
plt.title("Boxplot of Group 1 and Group 2")
plt.ylabel("Values")
plt.show()
```

# Q-27)  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(0)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)

observed_counts = np.array([np.sum(observed == i) for i in range(1, 7)])
expected_counts = np.array([1000 / 6] * 6)

chi2_stat, p_value = stats.chisquare(observed_counts, expected_counts)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies do not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies follow the expected distribution.")

categories = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(10, 6))
plt.bar(categories, observed_counts, alpha=0.7, label='Observed', color='blue')
plt.plot(categories, expected_counts, label='Expected', color='red', marker='o', linestyle='dashed', linewidth=2)
plt.xlabel('Categories (Die Faces)')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit Test: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
