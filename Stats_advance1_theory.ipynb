# Q-1) What is a random variable in probability theory.

Ans) Random variable in probability:

-> A random variable in probability theory is a numerical outcome of a random experiment.

-> It assigns a number to each possible outcome of an experiment.

**Types of Random Variables:**

* Discrete Random Variable

-> Takes a finite or countable number of values.

Example: Number of heads in 3 coin tosses (values: 0, 1, 2, 3).

* Continuous Random Variable

-> Takes an infinite number of values within a range.

Example: Height of students in a class (values: any real number in a range like 150 cm to 200 cm).

1) Discrete random varibale

# Q-18) What is the significance of confidence intervals in statistical analysis?

Ans) A confidence interval (CI) is a range of values used to estimate an unknown population parameter (such as a population mean or proportion) based on sample data.

-> The interval is accompanied by a confidence level that tells us how confident we are that the true parameter lies within that interval.

-> e.g)  A 95% confidence interval means that if you were to take many samples and calculate the interval for each, 95% of those intervals would contain the true population parameter.

-> Confidence intervals provide a range of plausible values for the population parameter, which is much more informative than just a single point estimate.

-> By giving a range, the CI helps to express the uncertainty or variability inherent in using sample data to estimate a population parameter.

-> Confidence intervals show the uncertainty of an estimate. A narrow interval suggests more precision, while a wider interval indicates less precision.

-> Confidence intervals help decision-makers understand the range of possible outcomes.

# Q-21) What are the assumptions for applying the Central Limit Theorem?

Ans) The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that, under certain conditions.

-> The sampling distribution of the sample mean will be approximately normal, regardless of the shape of the population distribution.

->  The samples must be selected in a way that each individual data point has an equal chance of being included in the sample.

->  This ensures that the sample is representative of the population, and the sampling process is unbiased.

-> One observation should not influence another. For example, the outcome of one trial in a survey should not affect the outcome of the next trial.

-> In the context of sampling, this means that the samples should be selected in such a way that they do not interfere with each other, i.e., no dependency between the observations

-> As the sample size increases, the sampling distribution of the sample mean approaches a normal distribution more closely.

-> The larger the sample size, the more accurate the approximation to a normal distribution will be.

# Q-22) What is the concept of expected value in a probability distribution?

Ans) The expected value (also called the mean or expectation) is a key concept in probability theory.

->  It represents the average outcome of a random variable, considering the probability of each possible outcome.

->  In simpler terms, it's the "long-run" average if you were to repeat an experiment or random process many times.

-> The expected value is the weighted average of all possible values a random variable can take, where each value is weighted by its probability of occurring.

-> Discrete Case: Sum the possible values weighted by their probabilities.

-> Continuous Case: Integrate the product of the variable and its probability density function.

-> Interpretation: It represents the long-run average or balance point of a distribution.

-> Applications: It is used in decision-making, risk management, game theory, and predictive modeling.

Ans) A probability distribution describes how the probabilities of different outcomes are distributed for a random variable.

-> The expected outcome (or expected value) is derived directly from this distribution and provides a summary measure that describes the central tendency of the distribution.

-> The expected value uses this distribution to calculate a weighted average, representing the long-run average outcome of the random variable.

-> The expected value serves as the center or balance point of the distribution, summarizing the distribution in a single value.
