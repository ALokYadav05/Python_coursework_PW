# Q-1) What is Simple Linear Regression?

Ans) Simple Linear Regression is a basic type of predictive analysis that helps us understand the relationship between two variables.

* Independent Variable (X) – The variable we use to make predictions.

* Dependent Variable (Y) – The variable we want to predict.

-> It assumes that there is a linear relationship between these two variables, meaning that if we plot them on a graph, the points will mostly form a straight-line trend.


The relationship between X and Y is represented by this equation:

* Y = M*x + c

# Q-2) What are the key assumptions of Simple Linear Regression?

Ans) For Simple Linear Regression to work properly and give reliable results, certain assumptions must be met.

-> These assumptions ensure that the model accurately captures the relationship between the independent variable (X) and the dependent variable (Y).

-> The relationship between X (independent variable) and Y (dependent variable) should be linear (i.e., a straight-line trend).

-> If the relationship is not linear, the predictions will not be accurate.

-> The data points should be independent of each other.

-> This means that one observation should not be influenced by the previous one.

-> The errors (differences between actual and predicted Y values) should follow a normal distribution.

-> This assumption ensures that the model provides the most accurate predictions.

-> Since Simple Linear Regression has only one independent variable, this assumption is not relevant.

# Q-5) How do we calculate the slope m in Simple Linear Regression?

Ans) In Simple Linear Regression, the slope
𝑚
m represents the rate of change of
Y with respect to
X. It tells us how much
Y changes when
X increases by 1 unit.

      

```python
import numpy as np

# Given data
X = np.array([2, 4, 6, 8, 10])
Y = np.array([50, 60, 65, 80, 90])

# Calculate means
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Compute slope (m)
numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)
m = numerator / denominator

print("Slope (m):", m)
```

# Q-6) What is the purpose of the least squares method in Simple Linear Regression?

Ans) The Least Squares Method is used in Simple Linear Regression to find the best-fitting straight line for a given dataset.

-> Avoids negatives: Squaring removes negative values so that positive and negative errors don’t cancel out.

-> Gives more weight to large errors: A large mistake is penalized more than a small one.

-> It minimizes the sum of squared errors (SSE), ensuring the most accurate predictions.

-> It provides the optimal values of
m (slope) and
c (intercept).

# Q-11) What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Ans) Heteroscedasticity refers to a situation in Multiple Linear Regression where the variance of the residuals (errors) is not constant across all levels of the independent variables.

-> In an ideal regression model, the spread of errors should be even across all values of
X (this is called homoscedasticity).

-> If the errors increase or decrease systematically, we have heteroscedasticity.

**Effects of Heteroscedasticity on Regression Results:**

* Biased Standard Errors → Affects the confidence intervals and makes hypothesis tests unreliable.

* Incorrect p-values → Variables may appear significant or insignificant incorrectly.

* Inefficient Estimates → The regression model still gives unbiased coefficients, but they may not be the best estimates (not the lowest variance).

* Poor Predictions → The model may perform poorly on new data due to unstable variance.

# Q-12) How can you improve a Multiple Linear Regression model with high multicollinearity?

Ans) Multicollinearity occurs when two or more independent variables (predictors) in a Multiple Linear Regression model are highly correlated.

-> This makes it difficult to determine the effect of each predictor on the dependent variable, leading to unstable coefficients and misleading p-values.

*  Remove Highly Correlated Variables:

-> If two or more variables provide similar information, remove one.
Use correlation matrix to decide which variable to drop.
Example:
If "Years of Experience" and "Age" are highly correlated, keep only one.

* Use Principal Component Analysis (PCA)

-> PCA transforms correlated variables into a smaller set of uncorrelated components.

-> Helps reduce dimensionality while keeping important information.
Downside: The transformed variables lose interpretability.

*  Use Ridge or Lasso Regression:

-> Ridge Regression (L2 Regularization)
Reduces coefficients, but keeps all variables.

-> Lasso Regression (L1 Regularization)
Eliminates less important variables (sets some coefficients to 0).

-> Elastic Net combines both Ridge and Lasso.

# Q-19) How would you interpret a large standard error for a regression coefficient?

Ans) In regression analysis, the standard error (SE) of a coefficient measures the precision of the estimated coefficient.

-> A large standard error indicates that the estimate is unstable, meaning it has high variability across different samples.

-> A large standard error indicates low confidence in the coefficient estimate, which may be due to:

-> Small sample size

-> Multicollinearity

-> High data variability

-> Weak predictor effect

* Best Practices:

-> Check p-values and t-statistics.

-> Use VIF to detect multicollinearity.

-> Increase sample size or collect better data.

-> Consider feature engineering or transformation.

# Q-20)  How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Ans) Heteroscedasticity occurs when the variance of residuals (errors) is not constant across all levels of the independent variable(s).

->  This violates a key assumption of Ordinary Least Squares (OLS) regression, which assumes homoscedasticity (constant variance of errors).

-> How to Identify: Look for fanning-out patterns in residual plots.

->  Why It Matters: It biases standard errors, making p-values & confidence intervals unreliable.

->  How to Fix: Use log transformation, robust SE, WLS, or add missing variables.

# Q-28) What are the limitations of polynomial regression?

Ans) While Polynomial Regression is useful for modeling non-linear relationships, it has several limitations that can affect its effectiveness.

*  Overfitting (High Variance):

-> High-degree polynomials can fit the training data too well, capturing noise instead of actual patterns.

-> This leads to poor generalization to new data (high variance).

*  Extrapolation is Unreliable :

-> Polynomial functions can behave unpredictably outside the given data range.

-> Even small changes in input can cause huge deviations.


* Harder Interpretation

-> Higher-degree polynomial terms are difficult to interpret compared to linear models.

-> Interaction effects make it hard to understand which feature influences the output.

->  Polynomial Regression is useful for modeling non-linear relationships, but it has drawbacks.

->  Overfitting, computational cost, and difficulty in interpretation are key limitations.

->  Use regularization, cross-validation, and feature selection to overcome these challenges.

Ans) Polynomial regression is implemented using Scikit-Learn’s PolynomialFeatures and LinearRegression.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


np.random.seed(42)
X = np.linspace(1, 10, 20).reshape(-1, 1)
Y = 3*X**2 + 2*X + np.random.randn(20, 1)*10

plt.scatter(X, Y, color='blue', label="Actual Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Data")
plt.legend()
plt.show()
```

```python
degree = 2  # Choose polynomial degree

# Transform X into polynomial features (X, X²)
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# View transformed features
print(f"Original X:\n{X[:5]}")
print(f"Polynomial Features (X, X²):\n{X_poly[:5]}")


# Train the model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict on training data
Y_pred = model.predict(X_poly)


# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, Y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Print model coefficients
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")


# Generate smooth X values for plotting
X_range = np.linspace(1, 10, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_range_pred = model.predict(X_range_poly)

# Plot data and regression curve
plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X_range, Y_range_pred, color='red', label=f"Polynomial Fit (Degree {degree})")
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Polynomial Regression (Degree {degree})")
plt.legend()
plt.show()
```
