# Q-2) Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

Ans)

# Q-3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

Ans)

```python
def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return (p ** x) * ((1 - p) ** (1 - x))

p = 0.7
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

# Q-6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?


Ans)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 10  # Range [low, high]

x = np.arange(low, high + 1)
cdf_values = randint.cdf(x, low, high + 1)

plt.step(x, cdf_values, where='mid', color='blue', label="CDF", linewidth=2)

plt.xlabel('Random Variable (x)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) - Discrete Uniform')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, high - low + 2))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```

# Q-9)  Write a Python function to calculate Z-scores from a dataset and plot them.

Ans)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev
    return z_scores

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

z_scores = calculate_z_scores(data)

sns.histplot(z_scores, bins=30, kde=True, color='blue', edgecolor='black', stat="density", label="Z-score Distribution")

plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Q-10)  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

Ans)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

low, high = 0, 10
sample_size = 30
num_experiments = 1000

def central_limit_theorem(low, high, sample_size, num_experiments):
    means = []
    for _ in range(num_experiments):
        sample = np.random.uniform(low, high, sample_size)
        means.append(np.mean(sample))
    return means

means = central_limit_theorem(low, high, sample_size, num_experiments)

sns.histplot(means, bins=30, kde=True, color='blue', edgecolor='black', stat="density", label="Sample Means")

plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem for Uniform Distribution")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Q-15) Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

# Q-16)  Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).


Ans)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    x = np.linspace(-5, 5, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    sns.lineplot(x=x, y=y, color='blue', label="Standard Normal Distribution")

    plt.xlabel("X")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (Mean=0, Std=1)")
    plt.axvline(x=0, color='red', linestyle='--', label="Mean (X=0)")  # Mark the mean at 0
    plt.legend()
    plt.grid(True)

    plt.show()

plot_standard_normal_distribution()
```

# Q-18) Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

Ans)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_z_score(X, mean, std_dev):
    return (X - mean) / std_dev

data = [12, 15, 14, 10, 13, 15, 16, 17, 18, 12, 13, 14, 11, 19, 20]
mean = np.mean(data)
std_dev = np.std(data)

X = 16

z_score = calculate_z_score(X, mean, std_dev)

print(f"Z-score for X = {X}: {z_score:.2f}")

cdf_value = stats.norm.cdf(z_score)

print(f"Cumulative probability for Z-score {z_score:.2f}: {cdf_value:.4f}")

x_vals = np.linspace(-4, 4, 1000)
y_vals = stats.norm.pdf(x_vals)

plt.plot(x_vals, y_vals, label='Standard Normal Distribution (Mean=0, Std=1)')
plt.axvline(x=z_score, color='r', linestyle='--', label=f'Z-score = {z_score:.2f}')

plt.fill_between(x_vals, y_vals, where=(x_vals <= z_score), color='red', alpha=0.3)

plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.title('Comparison of Z-score with Standard Normal Distribution')
plt.legend()

plt.grid(True)
plt.show()
```

# Q-19)  Implement hypothesis testing using Z-statistics for a sample dataset.

Ans)

```python
import numpy as np
import scipy.stats as stats

sample_data = [102, 98, 101, 104, 107, 99, 96, 105, 100, 103]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

population_mean = 100
population_std_dev = 5

alpha = 0.05

z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

critical_z_value = stats.norm.ppf(1 - alpha / 2)

if abs(z_statistic) > critical_z_value:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"Critical Z-value: {critical_z_value:.2f}")
print(f"Result: {result}")
```

# Q-20) Create a confidence interval for a dataset using Python and interpret the result.

Ans)

```python
import numpy as np
import scipy.stats as stats


sample_data = [102, 98, 101, 104, 107, 99, 96, 105, 100, 103]

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level

z_value = stats.norm.ppf(1 - alpha / 2)

margin_of_error = z_value * (sample_std_dev / np.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"{confidence_level*100}% Confidence Interval: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}")
```

# Q-23)  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

Ans)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lambda_param = 3

x_vals = np.arange(0, 11)
cdf_vals = stats.poisson.cdf(x_vals, lambda_param)

plt.step(x_vals, cdf_vals, where='post', label=f'Poisson CDF\n(lambda={lambda_param})', color='blue')

plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (={lambda_param})')
plt.legend()

plt.grid(True)
plt.show()

for x, cdf in zip(x_vals, cdf_vals):
    print(f"P(X <= {x}) = {cdf:.4f}")
```

# Q-27)  Implement Z-score normalization on a dataset and visualize its transformation.

Ans)

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 10
size = 1000

dataset = np.random.normal(mean, std_dev, size)
dataset_normalized = (dataset - np.mean(dataset)) / np.std(dataset)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(dataset, bins=30, color='blue', alpha=0.7)
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(dataset_normalized, bins=30, color='orange', alpha=0.7)
plt.title('Normalized Dataset (Z-score)')
plt.xlabel('Z-score Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

print(f"Original Dataset: Mean = {np.mean(dataset):.2f}, Std Dev = {np.std(dataset):.2f}")
print(f"Normalized Dataset: Mean = {np.mean(dataset_normalized):.2f}, Std Dev = {np.std(dataset_normalized):.2f}")
```

# Q-28)  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

Ans)

```python
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(dataset):
    skewness = stats.skew(dataset)
    kurtosis = stats.kurtosis(dataset)

    return skewness, kurtosis

mean = 50
std_dev = 10
size = 1000

dataset = np.random.normal(mean, std_dev, size)

skewness, kurtosis = calculate_skewness_kurtosis(dataset)

print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurtosis:.4f}")
```
